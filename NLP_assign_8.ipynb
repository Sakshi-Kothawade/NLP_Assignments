```python
import nltk
from nltk.wsd import lesk
from nltk.corpus import wordnet as wn
from nltk.tokenize import word_tokenize

# Function for Word Sense Disambiguation
def perform_wsd(sentence, ambiguous_word):
    tokens = word_tokenize(sentence)

    # Apply Lesk Algorithm
    sense = lesk(tokens, ambiguous_word)

    if sense:
        print("\nWord:", ambiguous_word)
        print("Predicted Sense:", sense.name())
        print("Definition:", sense.definition())
        print("Example:", sense.examples())
    else:
        print("No sense found.")

# Example Sentences

sentence1 = "I deposited money in the bank."
sentence2 = "The river overflowed the bank after heavy rain."

print("Sentence 1:", sentence1)
perform_wsd(sentence1, "bank")

print("\n-----------------------------")

print("Sentence 2:", sentence2)
perform_wsd(sentence2, "bank")
```
