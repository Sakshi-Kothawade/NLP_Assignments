```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch
import torch.nn.functional as F

# Load multilingual sentiment model
model_name = "nlptown/bert-base-multilingual-uncased-sentiment"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)

# Sentiment Prediction Function
def analyze_sentiment(text):
    inputs = tokenizer(text, return_tensors="pt", truncation=True)
    outputs = model(**inputs)

    probs = F.softmax(outputs.logits, dim=1)
    predicted_class = torch.argmax(probs).item() + 1

    print("\nText:", text)
    print("Predicted Sentiment (1-5 stars):", predicted_class)
    print("Confidence:", torch.max(probs).item())

# Example Hindi Sentences
sentence1 = "यह फिल्म बहुत शानदार है।"
sentence2 = "मुझे यह सेवा बिल्कुल पसंद नहीं आई।"
sentence3 = "यह अनुभव ठीक था।"

analyze_sentiment(sentence1)
analyze_sentiment(sentence2)
analyze_sentiment(sentence3)
```

```python
def analyze_sentiment_label(text):
    inputs = tokenizer(text, return_tensors="pt", truncation=True)
    outputs = model(**inputs)

    probs = F.softmax(outputs.logits, dim=1)
    predicted_class = torch.argmax(probs).item() + 1

    if predicted_class <= 2:
        sentiment = "Negative"
    elif predicted_class == 3:
        sentiment = "Neutral"
    else:
        sentiment = "Positive"

    print("\nText:", text)
    print("Sentiment:", sentiment)
    print("Confidence:", torch.max(probs).item())
```

```python
while True:
    text = input("\nEnter Hindi sentence (or type 'exit'): ")
    if text.lower() == "exit":
        break
    analyze_sentiment_label(text)
```
